```python
# Step 1: Load the dataset
file_path = '/content/UNdata_Export_20250422_215048149.csv'
df = pd.read_csv(file_path)
```

```python
# Step 2: Define keywords for different categories
consumable_keywords = ['Food', 'Beverage', 'Tobacco', 'dairy', 'meat', 'grain']
textile_keywords = ['Textile', 'Clothing', 'Apparel', 'Garment', 'Fabric', 'Footwear', 'Leather']
electronics_keywords = ['Electric', 'Electronic', 'Appliance', 'Computer', 'Instruments', 'TV', 'Radio', 'Telecom', 'Circuit', 'Machinery', 'Wire', 'Batteries']
transport_keywords = ['Transport', 'Vehicle', 'Automobile', 'Railway', 'Engine', 'Trailer', 'Ship']
```

```python
# Step 3: Updated classification function
def classify_industry(industry_name):
    industry_lower = industry_name.lower()

    if any(word.lower() in industry_lower for word in transport_keywords):
        return 'Transport and Vehicles'
    elif any(word.lower() in industry_lower for word in electronics_keywords):
        return 'Machinery, Electrics and Electronics'
    elif any(word.lower() in industry_lower for word in textile_keywords):
        return 'Textiles and Clothing'
    elif any(word.lower() in industry_lower for word in consumable_keywords):
        return 'Food Industry'
    else:
        return 'Other Industrial Items'
```

```python
# Step 4: Apply classification
df['Industry_Category'] = df['ISIC Rev 3'].apply(classify_industry)
```

```python
# Step 5: Drop unwanted columns
columns_to_drop = ['Measure', 'Value Footnotes']
for col in columns_to_drop:
    if col in df.columns:
        df = df.drop(columns=[col])
```

```python
if 'Value' in df.columns:
    df = df.rename(columns={'Value': 'Number of Employees'})
```

```python
# Step 6: Save the updated DataFrame to a new CSV
output_path = '/content/processed_industry_data.csv'
df.to_csv(output_path, index=False)
```
